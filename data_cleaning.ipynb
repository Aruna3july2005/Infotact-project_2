```python

import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/PROJECT_2/hotel_bookings.csv')

print("=" * 60)
print("BEFORE CLEANING")
print("=" * 60)
print(f"Shape          : {df.shape}")
print(f"Total nulls    : {df.isnull().sum().sum()}")
print(f"\nMissing per column (only columns with nulls):")
print(df.isnull().sum()[df.isnull().sum() > 0])

```

```python
df['children'] = df['children'].fillna(0).astype(int)
print("\n FIX 1 | children: filled 4 nulls with 0")
```

```python
df['agent'] = df['agent'].fillna(0).astype(int)
print(" FIX 2 | agent: filled 16,340 nulls with 0 (No Agent)")
```

```python
df.drop(columns=['company'], inplace=True)
print(" FIX 3 | company: dropped (94% missing — not useful)")
```

```python
most_common_country = df['country'].mode()[0]
df['country'] = df['country'].fillna(most_common_country)
print(f" FIX 4 | country: filled 488 nulls with mode → '{most_common_country}'")
```

```python
before = len(df)
df = df[df['adr'] >= 0]          # remove negative
df = df[df['adr'] <= 1000]       # remove extreme high outliers
after = len(df)
print(f" FIX 5 | adr: removed {before - after} outlier rows "
      f"(negative or > 1000). Rows remaining: {after}")
```

```python
ghost = df[(df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0)]
df = df.drop(ghost.index)
print(f" BONUS | Dropped {len(ghost)} ghost bookings (0 adults + 0 children + 0 babies)")
```

```python
zero_stay = df[(df['stays_in_weekend_nights'] == 0) & (df['stays_in_week_nights'] == 0)]
df = df.drop(zero_stay.index)
print(f" BONUS | Dropped {len(zero_stay)} bookings with 0 total nights")
```
