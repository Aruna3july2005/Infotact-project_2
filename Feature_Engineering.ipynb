```python
import numpy as np
import pandas as pd
```

```python
df['total_stay'] = (
    df['stays_in_weekend_nights'] +
    df['stays_in_week_nights']
)
```

```python
df['arrival_month_num'] = pd.to_datetime(
    df['arrival_date_month'],
    format='%B'
).dt.month
```

```python
df['arrival_date'] = pd.to_datetime({
    'year': df['arrival_date_year'],
    'month': df['arrival_month_num'],
    'day': df['arrival_date_day_of_month']
})
```

```python
df['children'] = df['children'].fillna(0)

df['is_family'] = np.where(
    (df['adults'] > 1) &
    ((df['children'] > 0) | (df['babies'] > 0)),
    1,
    0
)
```

```python
df['revenue'] = np.where(
    df['is_canceled'] == 0,
    df['adr'] * df['total_stay'],
    0
)
```

```python
df['booker_type'] = np.where(
    df['lead_time'] > 90,
    'Early Planner',
    np.where(
        df['lead_time'] >= 30,
        'Mid',
        'Last Minute'
    )
)
```
